# Morphological Transformations

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# loading an orange image
image = cv2.imread('sad.jpg',0)
```

```python
kernel = np.ones((5,5),np.uint8)
image_erosion = cv2.erode(image,kernel,iterations = 3)
image_dilation = cv2.dilate(image,kernel,iterations = 3)
```

```python
plt.subplot(131),plt.imshow(image,cmap='gray'),plt.title('original')
plt.subplot(132),plt.imshow(image_erosion,cmap='gray'),plt.title('erosion')
plt.subplot(133),plt.imshow(image_dilation,cmap='gray'),plt.title('dilation')
```

```python
# loading an orange image
imageBGR = cv2.imread('orange.jpg',-1)
# convert the image from BGR color space to RGB
imageRGB=cv2.cvtColor(imageBGR, cv2.COLOR_BGR2RGB)

# Convert BGR to HSV
imageHSV = cv2.cvtColor(imageBGR, cv2.COLOR_BGR2HSV)

# define range of orange color in HSV
lower_orange = np.array([0,200,200])
upper_orange = np.array([40,255,255])


# Threshold the HSV image to get only orange colors
mask = cv2.inRange(imageHSV, lower_orange, upper_orange)



# Bitwise-AND mask and original image
res = cv2.bitwise_and(imageRGB,imageRGB, mask= mask)

plt.figure(figsize=(15,15))
ax = plt.subplot2grid((1, 3), (0, 0))
ax.imshow(imageRGB)
plt.title('RGB')

ax = plt.subplot2grid((1, 3), (0, 1))
ax.imshow(mask,cmap='gray')
plt.title('mask')

ax = plt.subplot2grid((1, 3), (0, 2))
ax.imshow(res)
plt.title('Result')
```

```python
#opening
kernel = np.ones((6,6),np.uint8)
modified_mask= cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel, iterations = 0)
modified_mask = cv2.dilate(modified_mask,kernel,iterations = 12)
modified_mask = cv2.erode(modified_mask,kernel,iterations = 16)


# Bitwise-AND mask and original image
res = cv2.bitwise_and(imageRGB,imageRGB, mask= modified_mask)

plt.figure(figsize=(15,15))
ax = plt.subplot2grid((1, 3), (0, 0))
ax.imshow(imageRGB)
plt.title('RGB')

ax = plt.subplot2grid((1, 3), (0, 1))
ax.imshow(modified_mask,cmap='gray')
plt.title('mask')

ax = plt.subplot2grid((1, 3), (0, 2))
ax.imshow(res)
plt.title('Result')
```
