# Contours

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# loading an orange image
imageBGR = cv2.imread('sign.png',-1)
# convert the image from BGR color space to RGB
imageRGB=cv2.cvtColor(imageBGR, cv2.COLOR_BGR2RGB)



# Convert BGR to HSV
imageHSV = cv2.cvtColor(imageBGR, cv2.COLOR_BGR2HSV)

# define range of orange color in HSV
lower_orange = np.array([20,160,80])
upper_orange = np.array([55,255,255])


# Threshold the HSV image to get only orange colors
mask = cv2.inRange(imageHSV, lower_orange, upper_orange)


# Bitwise-AND mask and original image
res = cv2.bitwise_and(imageRGB,imageRGB, mask= mask)

plt.figure(figsize=(15,15))
ax = plt.subplot2grid((1, 3), (0, 0))
ax.imshow(imageRGB)
plt.title('RGB')

ax = plt.subplot2grid((1, 3), (0, 1))
ax.imshow(mask,cmap='gray')
plt.title('mask')

ax = plt.subplot2grid((1, 3), (0, 2))
ax.imshow(res)
plt.title('Result')
```

```python
_, contours, hierarchy = cv2.findContours(mask,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
imagencountour=imageRGB.copy()
cv2.drawContours(imagencountour, contours, 3, (200,50,40), 10);
```

```python
plt.figure(figsize=(15,15));
plt.subplot(131),plt.imshow(imageRGB),plt.title('imageRGB');
plt.subplot(132),plt.imshow(mask,cmap='gray'),plt.title('mask');
plt.subplot(133),plt.imshow(imagencountour),plt.title('contoured');
```

```python
cv2.drawContours(imagencountour, contours, 1, (10,250,20), 10);

plt.figure(figsize=(35,35));
plt.subplot(121),plt.imshow(imageRGB),plt.title('imageRGB');
plt.subplot(122),plt.imshow(imagencountour),plt.title('contoured');
```
