# Transformations

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# loading an orange image
imageBGR = cv2.imread('car.jpg',-1)
# convert the image from BGR color space to RGB
imageRGB=cv2.cvtColor(imageBGR, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(imageRGB)
imageRGB.shape
```

```python
#lets ressize it
image_resized = cv2.resize(imageRGB,None,fx=2, fy=2, interpolation = cv2.INTER_CUBIC)
image_resized.shape
plt.imshow(image_resized)
```

```python
#rotate the original image
rows,cols,ch = imageRGB.shape
M = cv2.getRotationMatrix2D((cols/2,rows/2),40,1)
image_rotated = cv2.warpAffine(imageRGB,M,(cols,rows))
plt.imshow(image_rotated)
```

```python
# affine transformation
rows,cols,ch = imageRGB.shape

pts1 = np.float32([[20,50],[100,50],[50,200]])
pts2 = np.float32([[10,150],[100,50],[100,250]])

M = cv2.getAffineTransform(pts1,pts2)

image_affine = cv2.warpAffine(imageRGB,M,(cols,rows))

plt.subplot(121),plt.imshow(imageRGB),plt.title('Input')
plt.subplot(122),plt.imshow(image_affine),plt.title('Output')
plt.show()
```
