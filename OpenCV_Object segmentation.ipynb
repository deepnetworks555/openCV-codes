# Object segmentation based on color

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# loading an orange image
imageBGR = cv2.imread('orange.jpg',-1)
# convert the image from BGR color space to RGB
imageRGB=cv2.cvtColor(imageBGR, cv2.COLOR_BGR2RGB)
```

```python
ax = plt.subplot2grid((1, 2), (0, 0))
ax.imshow(imageRGB)
plt.title('RGB')
ax = plt.subplot2grid((1, 2), (0, 1))
ax.imshow(imageBGR)
plt.title('BGR')
```

```python
# Convert BGR to HSV
imageHSV = cv2.cvtColor(imageBGR, cv2.COLOR_BGR2HSV)

# define range of orange color in HSV
lower_orange = np.array([0,160,80])
upper_orange = np.array([25,255,255])


# Threshold the HSV image to get only orange colors
mask = cv2.inRange(imageHSV, lower_orange, upper_orange)
```

```python
# Bitwise-AND mask and original image
res = cv2.bitwise_and(imageRGB,imageRGB, mask= mask)
```

```python
ax = plt.subplot2grid((1, 3), (0, 0))
ax.imshow(imageRGB)
plt.title('RGB')

ax = plt.subplot2grid((1, 3), (0, 1))
ax.imshow(mask,cmap='gray')
plt.title('mask')

ax = plt.subplot2grid((1, 3), (0, 2))
ax.imshow(res)
plt.title('Result')
```
